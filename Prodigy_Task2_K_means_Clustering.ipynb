```python
# Importing the dependecies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Importing the dataset
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
sns.countplot(data=df, x='Gender')
```

```python
X = df.iloc[:,[3,4]]
X
```

```python
# using the elbow method to determine the number of clusters
#using elbow method to determine the optimal number of clusters we can use. We use WCSS values
wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters = i,init='k-means++' ,random_state=0, n_init = "auto")
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS values')
```

```python
#The Model
model = KMeans(n_clusters = 5, init = "k-means++", random_state = 0)
```

```python
y_pred = model.fit_predict(X)
y_pred
```

```python
centroids = model.cluster_centers_
centroids
```

```python
df["Cluster"] = y_pred
sns.scatterplot(x="Annual Income (k$)",y= "Spending Score (1-100)", data = df, hue="Cluster", palette=['green','orange','Purple','dodgerblue','violet'], legend='full')
plt.scatter(centroids[:,0], centroids[:,1], color = "Black")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Predicted Clusters')
plt.show()
```
